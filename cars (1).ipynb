#### What are the different types of car brand and how many times did each type occur

```python
#displaying different types of car brand and their occurence

car_make_count = cars['Make'].value_counts()
car_make_count
```

```python
plt.figure(figsize=(10,5))
cars['Make'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Car Brand')
plt.ylabel('Count')
plt.title('Number of car by brand')
plt.show()
```

```python
cars['Origin'].unique()
```

```python
cars['Origin'].value_counts().plot(kind = 'bar', color = 'skyblue')
plt.title('Number of Cars by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')
plt.show()
```

```python
cars_asia = cars[cars['Origin'].isin(['Asia'])]
count_asia = cars_asia['Make'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(count_asia.index, count_asia.values, color='skyblue')
plt.xlabel('Car Make (Brand)')
plt.ylabel('Count')
plt.title('Car Brands from Asia')
plt.xticks(rotation=90)
plt.show()
```

```python
cars_asia = cars[cars['Origin'].isin(['Europe'])]
count_asia = cars_asia['Make'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(count_asia.index, count_asia.values, color='skyblue')
plt.xlabel('Car Make (Brand)')
plt.ylabel('Count')
plt.title('Car Brans from Europe')
plt.xticks(rotation=90) 
plt.show()
```

```python
cars_asia = cars[cars['Origin'].isin(['USA'])]
count_asia = cars_asia['Make'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(count_asia.index, count_asia.values, color='skyblue')
plt.xlabel('Car Make (Brand)')
plt.ylabel('Count')
plt.title('Car Brands from USA')
plt.xticks(rotation=90) 
plt.show()
```

```python
#creating new 'Price' column from column 'Invoice'
cars['Price'] = cars['Invoice'].str.replace('$','')

# Remove non-numeric characters from 'Price' column
cars['Price'] = cars['Price'].str.replace(",","").replace(" ","")

# Convert 'Price' column to integer
cars['Price'] = cars['Price'].astype(int)

cars.head()
```

```python
cars.dtypes
```

```python
# Create a figure with a larger size
plt.figure(figsize=(12, 8))

# Filter cars from Asia and Europe
filtered_cars = cars[cars['Origin'].isin(['Asia', 'Europe'])]

# Calculate the average price for cars from Asia and Europe
avg_price = filtered_cars['Price'].mean()

# Create a scatter plot for cars from Asia and Europe
sns.scatterplot(x='Make', y='Price', data=filtered_cars, hue='Origin', color='skyblue', marker='o', s=100)

# Add a horizontal line to indicate the average price
plt.axhline(y=avg_price, color='black', linestyle='dashed', label=f'Average Price: ${avg_price:.2f}')

# Set labels and title
plt.title('Average Car Prices by Make (Asia and Europe)', fontsize=20)
plt.xlabel('Make', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.legend(loc='upper left', fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Add grid lines
plt.grid()

plt.tight_layout()  # Ensure labels are not cut off
plt.show()
```

```python
# Sort the dataframe by Price in descending order
cars = cars.sort_values(by='Price',ascending=False)

# Select the top 15 car types
cars = cars.head(15)

# Create a bar plot
sns.barplot(x = 'Make', y = 'Price', data = cars, hue = 'Origin',palette='Blues')

# Set labels and titles
plt.title("Price Distribution of Top 7 Car Types by Make and Origin", fontsize=20)
plt.xlabel("Make",fontsize=14)
plt.ylabel("Price",fontsize=14)
plt.xticks(rotation=90)
plt.legend(loc='upper left',fontsize=14)
plt.grid()
plt.show()
```

```python
#Plot size
plt.figure(figsize=(14,8))

#create a scatter plot of engine size vs horsepower
sns.scatterplot(x='EngineSize', y='Horsepower', data=cars, hue='Origin', palette = 'Purples')

#set the title, labels and size of the title
plt.title("Relation between Engine Size and Horsepower by Origin", fontsize=15)
plt.xlabel("Engine Size", fontsize=12)
plt.ylabel("Horsepower", fontsize=12)
plt.legend(loc='upper left') #legend location
plt.show()
```

```python
# Group the data by car type and calculate the average cylinder size
car_data_grouped = cars.groupby('Make')['Cylinders'].mean().reset_index()

# Sort the data by average cylinder size in descending order
car_data_grouped.sort_values(by='Cylinders', ascending=False, inplace=True)

# Select the top 10 car types by average cylinder size
top_10_car_types = car_data_grouped.head(10)

# Plot size
plt.figure(figsize=(10, 5))

# Create a bar plot of the average cylinder size by top 10 car types
sns.barplot(x='Make', y='Cylinders', palette='PuBu', data=top_10_car_types)

# Set the title and labels
plt.title("Average Cylinder Size by Top 10 Car Makes", fontsize=16)
plt.xlabel("Car Make")
plt.ylabel("Average Cylinder Size")
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees for better readability
plt.show()
```

```python
#plot size
plt.figure(figsize=(20,10))

sns.lmplot(x='EngineSize', y='Horsepower', data=cars, hue='Origin',palette = 'YlOrBr', fit_reg=True)
#set the title and labels
plt.title("Relation between EngineSize and Horsepower by Origin", fontsize=16)
plt.xlabel("Engine Size")
plt.ylabel("Horsepower")
plt.show()
```
